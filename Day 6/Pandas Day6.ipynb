## Manipulating Data Part 2

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
#creating a column
seat_column = pd.Series([5,5,5,5,5,5])

#new column called seats adding it to the dataset
car_sales["Seats"] = seat_column
car_sales
```

```python
car_sales
```

```python
#Column from a Python List
fuel_economy = [7.6, 8.9, 9.0, 8.5, 3.0, 7.9, 5.5, 6.7, 9.9, 10]
car_sales["Fuel per 100KM"] = fuel_economy
```

```python
car_sales
```

```python
car_sales["Total Fuel (L)"] = car_sales["Odometer (KM)"] / 100 * car_sales["Fuel per 100KM"]
car_sales
```

```python
#removing a column
car_sales
car_sales = car_sales.drop("Seats", axis = 1)
```
